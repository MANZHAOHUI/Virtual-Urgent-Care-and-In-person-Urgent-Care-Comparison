```python
import numpy as np
import pandas as pd
from pandas import DataFrame,Series
def detect_outliers(df,col):
    
    Q1 = np.percentile(df[col],25)
    Q3 = np.percentile(df[col],75)
        # interquartile range(IQR)
    IQR = Q3 - Q1

    outlier_step = 1.5 * IQR

        # Determine a list of indeices of ouliers for feature col
    outlier_list_col = df[(df[col] < Q1 - outlier_step) | (df[col] > Q3 + outlier_step)].index.tolist()


    return outlier_list_col

df = pd.DataFrame(pd.read_csv('c:\data\Saif/INPERSON_TOP10_SERVICECOST_XL.csv'))
df.describe()
```

```python
outlier_list=detect_outliers(df, 'SERVICE_COST')
len(outlier_list)
```

```python
for row in outlier_list:
    print (row,df.iloc[row]['SERVICE_COST'])
    
```

```python
for row in outlier_list:
    df=df.drop([row])

df.shape
```
